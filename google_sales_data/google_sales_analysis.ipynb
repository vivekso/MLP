```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('GoogleAds_DataAnalytics_Sales_Uncleaned.csv')
df.head()
```

```python
df[df['Clicks'].isna()]
# df['Clicks_imputed'] = df['Clicks'].isnull()
# df[df['Clicks_imputed'] == False]
```

```python
# Estimate clicks from Conversions and Conversion Rate
df['Estimated_Clicks_1'] = df['Conversions']/ df['Conversion Rate']
# df
# Estimate from CTR (global average)
global_ctr = (df['Clicks']/df['Impressions']).mean()
# global_ctr
df['Estimated_Clicks_2'] = df['Impressions'] * global_ctr
df
# Use Estimated_Clicks_1 if valid, else fallback to Estimated_Clicks_2
df['Clicks_filled'] = df['Clicks']
df['Clicks_filled'] = df['Clicks_filled'].fillna(df['Estimated_Clicks_1'])
df['Clicks_filled'] = df['Clicks_filled'].fillna(df['Estimated_Clicks_2'])

# Clean up
df['Clicks'] = df['Clicks_filled'].round()
df.drop(columns=['Estimated_Clicks_1', 'Estimated_Clicks_2', 'Clicks_filled'], inplace=True)
df
```

```python
df['Cost'] = df['Cost'].replace('[\$]', '', regex=True).astype(float)
df['Sale_Amount'] = df['Sale_Amount'].replace('[\$]', '', regex=True).astype(float)

df
```

```python
# df['Ad_Date'] = pd.to_datetime(df['Ad_Date'], errors='coerce', dayfirst=True)
# df[df['Ad_Date'].isna()]
import re

def guess_format(s):
    s = str(s).strip()
    if re.match(r'^\d{2}/\d{2}/\d{4}$', s):  # 28/07/2023 (DD/MM/YYYY)
        return 'DD/MM/YYYY'
    elif re.match(r'^\d{4}-\d{2}-\d{2}$', s):  # 2023-07-28 (YYYY-MM-DD)
        return 'YYYY-MM-DD'
    elif re.match(r'^\d{2}-\d{2}-\d{4}$', s):  # 28-07-2023 (DD-MM-YYYY)
        return 'DD-MM-YYYY'
    elif re.match(r'^\d{4}/\d{2}/\d{2}$', s):  # 2024/11/22 (YYYY/MM/DD)
        return 'YYYY/MM/DD'
    else:
        return 'Unknown'
df['guessed_format'] = df['Ad_Date'].astype(str).apply(guess_format)
```

```python
df.loc[df['guessed_format'] == 'YYYY-MM-DD', 'parsed_date'] = pd.to_datetime(
    df.loc[df['guessed_format'] == 'YYYY-MM-DD', 'Ad_Date'], format='%Y-%m-%d', errors='coerce')
df.loc[df['guessed_format'] == 'DD-MM-YYYY', 'parsed_date'] = pd.to_datetime(
    df.loc[df['guessed_format'] == 'DD-MM-YYYY', 'Ad_Date'], format='%d-%m-%Y', errors='coerce')
df.loc[df['guessed_format'] == 'DD-MM-YYYY', 'parsed_date'] = pd.to_datetime(
    df.loc[df['guessed_format'] == 'DD-MM-YYYY', 'Ad_Date'], format='%Y-%m-%d', errors='coerce')
df.loc[df['guessed_format'] == 'YYYY/MM/DD', 'parsed_date'] = pd.to_datetime(
    df.loc[df['guessed_format'] == 'YYYY/MM/DD', 'Ad_Date'], format='%Y/%m/%d', errors='coerce')
df.loc[df['guessed_format'] == 'DD/MM/YYYY', 'parsed_date'] = pd.to_datetime(
    df.loc[df['guessed_format'] == 'DD/MM/YYYY', 'Ad_Date'], format='%d/%m/%Y', errors='coerce')
df.loc[df['guessed_format'] == 'Unknown', 'parsed_date'] = pd.NaT
df.loc[df['guessed_format'] == 'DD-MM-YYYY', 'parsed_date'] = pd.to_datetime(
    df.loc[df['guessed_format'] == 'DD-MM-YYYY', 'Ad_Date'], format='%d-%m-%Y', errors='coerce')
df
```
